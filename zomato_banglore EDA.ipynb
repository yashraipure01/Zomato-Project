# Cleaning rate Column

```python
df["rate"].unique()
```

## removing "NEW" , "-" , "/5" from rate Column

```python
def handlerate(value):
    if(value == "NEW" or value== "-"):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
```

```python
df["rate"] = df["rate"].apply(handlerate)
df["rate"].head()
```

## Filling null values in rate column

```python
df.rate.isnull().sum()
```

```python
df["rate"].fillna(df["rate"].mean() , inplace = True)
df["rate"].isnull().sum()
```

# Dropping Null values

```python
df.info()
```

```python
df.dropna(inplace = True)
df.info()
```

# Renaming columns for better understanding

```python
df.head(1)
```

```python
df.rename(columns = {"approx_cost(for two people)":"Cost2plates" ,"listed_in(type)": "Type"} , inplace = True)
df.head(2)
```

# location and listed_in(city) are almost same 
## so dropping one

```python
df = df.drop(["listed_in(city)"] , axis = 1 )
```

```python
df.head()
```

# Handeling cost2plates column

```python
df["Cost2plates"].unique()
```

## Some values have ",", which restrict them from becoming int

```python
def handlecomma(value):
    value = str(value)
    if "," in value:
        value = value.replace("," , "")
        return float(value)
    else:
        return float(value)
```

```python
df["Cost2plates"] = df["Cost2plates"].apply(handlecomma)
df["Cost2plates"].unique()
```

```python
df.head(3)
```

# Cleaning rest_types Column

```python
df["rest_type"].value_counts()
```

```python
rest_types = df["rest_type"].value_counts(ascending = False)
rest_types
```

## soring rest_types who has less than 1000 outlets

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

## Updating df as rest_types as others who has less than 1000 restaurent

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return "others"
    else:
        return value
```

```python
df["rest_type"] = df["rest_type"].apply(handle_rest_type)
df["rest_type"].value_counts()
```

# Cleaning location column

```python
df["location"].value_counts()
```

```python
location = df["location"].value_counts(ascending = False)
```

```python
location_lessthan300 = location[location<300]
```

## clustering locaton less than 300 as others

```python
def handle_location(value):
    if(value in location_lessthan300):
        return "others"
    else:
        return value
df["location"] = df["location"].apply(handle_location)
df["location"].value_counts()
```

```python
df.head(2)
```

# Analyzing cusines 

```python
cuisines = df["cuisines"].value_counts(ascending = False)
cusiness_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cusiness_lessthan100):
        return "othres"
    else:
        return value
```

```python
df["cuisines"] = df["cuisines"].apply(handle_cuisines)
df["cuisines"].value_counts()
```
